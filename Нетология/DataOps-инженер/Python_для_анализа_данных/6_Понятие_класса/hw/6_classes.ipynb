```python
"""
Задание 1
Напишите функцию, которая возвращает название валюты (поле ‘Name’)
с максимальным значением курса с помощью сервиса www.cbr-xml-daily.ru...ly_json.js

Задание 2
Добавьте в класс Rate параметр diff (со значениями True или False), который в случае значения True
в методах курсов валют (eur, usd итд) будет возвращать не курс валюты, а изменение по сравнению в прошлым значением.
Считайте, self.diff будет принимать значение True только при возврате значения курса.
При отображении всей информации о валюте он не используется.
"""
import requests


class Rate:
    def __init__(self):
        self.diff_ = False
        self.full_data = requests.get("https://www.cbr-xml-daily.ru/daily_json.js").json()

    def set_diff(self, is_diff):
        self.diff_ = is_diff

    def get_code_with_max_rate(self, in_data=None):
        if in_data is None:
            in_data = self.full_data
        max_val = []
        for k in in_data['Valute']:
            if len(max_val) == 0:
                max_val = [in_data['Valute'][k]['Name'], in_data['Valute'][k]['Value']]
            else:
                if max_val[1] < in_data['Valute'][k]['Value']:
                    max_val = [in_data['Valute'][k]['Name'], in_data['Valute'][k]['Value']]
        return max_val[0]

    def get_rate(self, in_code):
        if self.diff_:
            return self.full_data['Valute'][in_code]['Value'] - self.full_data['Valute'][in_code]['Previous']
        else:
            return self.full_data['Valute'][in_code]['Value']


# test
my_rate = Rate()
print(my_rate.get_code_with_max_rate())
my_rate.set_diff(False)
print(my_rate.get_rate('USD'))
my_rate.set_diff(True)
print(my_rate.get_rate('USD'))
```

```python
"""
Задание 3
Напишите класс Designer, который учитывает количество международных премий. Подсказки в коде занятия
в разделе “Домашнее задание задача 3”.

Комментарий по классу Designer такой:
Напишите класс Designer, который учитывает количество международных премий для дизайнеров
(из презентации: “Повышение на 1 грейд за каждые 7 баллов. Получение международной премии – это +2 балла”).
Считайте, что при выходе на работу сотрудник уже имеет две премии и их количество не меняется со стажем
(конечно если хотите это можно вручную менять).

Класс Designer пишется по аналогии с классом Developer из материалов занятия.
"""
class Designer:

    def __init__(self, name, seniority):
        self.name = name
        self.inter_award_point = 2
        self.seniority = seniority + self.inter_award_point

        self.grade = 1

    def grade_up(self):
        self.grade += 1

    def publish_grade(self):
        print(self.name, self.grade)

    def chk_if_it_is_the_time_for_up(self):
        self.seniority += 1

        if self.seniority % 7 == 0:
            self.grade_up()

        return self.publish_grade()


# test
des = Designer("Василий", 1)
for i in range(30):
    des.chk_if_it_is_the_time_for_up()
```
