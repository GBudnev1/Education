```python
'''
Переведите содержимое файла purchase_log.txt в словарь purchases вида:
{‘1840e0b9d4’: ‘Продукты’, …}

Для каждого user_id в файле visit_log.csv определите третий столбец с категорией покупки
(если покупка была, сам файл visit_log.csv изменять не надо). Запишите в файл funnel.csv
визиты из файла visit_log.csv, в которых были покупки с указанием категории.
Учтите условия на данные:
содержимое purchase_log.txt помещается в оперативную память компьютера
содержимое visit_log.csv - нет; используйте только построчную обработку этого файла
'''
import json
from pathlib import Path
import os


def run_main():
    """
    исходные файлы необходимо поместить в пользовательский каталог
    """
    f_pur = os.path.join(Path.home(), "purchase_log.txt")
    f_visits = os.path.join(Path.home(), "visit_log.csv")
    f_funnel = os.path.join(Path.home(), "funnel.csv")

    pur_dict = get_pur_dict(f_source=f_pur)
    merge_pur_and_visits(in_dict=pur_dict, f_source=f_visits, f_target=f_funnel)


def merge_pur_and_visits(in_dict, f_source, f_target):
    """
    Сопоставляем user_id из покупок с визитами и пишем инормацию из визитов с добавлением категории в funnel.csv
    :param in_dict:
    :param f_source:
    :param f_target:
    :return:
    """
    with open(f_source, 'r', encoding="utf-8") as visit_f:
        with open(f_target, "w", encoding="utf-8") as fun_f:
            fun_f.writelines(f"{visit_f.readline().strip()},category\n")
            for i, line in enumerate(visit_f):
                tmp_user_id = in_dict.get(line.split(',')[0])
                if tmp_user_id:
                    fun_f.writelines(f"{line.strip()},{','.join(in_dict.get(line.split(',')[0]))}\n")


def get_pur_dict(f_source):
    """
    Получаем словарь с данными из файла покупок
    :param f_source:
    :return:
    """
    pur_dict = {}
    with open(f_source, 'r', encoding="utf-8") as pur_f:
        next(pur_f)
        for i, line in enumerate(pur_f):
            line = line.strip()
            d = json.loads(line)
            if d['user_id'] in pur_dict.keys():
                pur_dict[d['user_id']] = list(set(pur_dict.get(d['user_id']) + [d['category']]))
            else:
                pur_dict[d['user_id']] = [d['category']]
    return pur_dict


if __name__ == '__main__':
    run_main()
```
